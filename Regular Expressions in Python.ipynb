# Regular Expressions in Python.

#### Regular expressions can be used to search for and match specific  text patterns of texts within strings. <br> You can create a regular expression for just about any kind of pattern you can think of.

```python
# Let's begin by importing the re module, for working with regular expressions in Python.
import re
```

```python
# Next we have a text to search....
text_to_search = '''
abcdefghijklmnopqrstuvwxyz
ABCDEFGHIJKLMNOPQRSTUVWXYZ
1234567890

Ha HaHa 

MetaCharacters (Need to be escaped):
. ^ $ * + ? { } [ ] \ | ( )

coreyms.com


321-555-4321
123.555.1234
123*555*1278
800.525.1234
900.545.1234
800.535.1234
900.515.1234

Mr. Schafer
Mr Smith
Ms Davis
Mrs. Robinson
Mr. T

_
cat
mat
pat
bat
'''

sentence = 'Start a sentence and then bring it to an end'
```

```python
# let's find say a pattern 'abc'
pattern = re.compile(r'abc')
```

```python
matches = pattern.finditer(text_to_search)
```

```python
for match in matches:
    print(match.)
```

```python
# the dot(.) is a special character in regular expressions. It actually matches everything.
pattern = re.compile(r'.')
matches = pattern.finditer(text_to_search)
for match in matches:
    print(match)
```

```python
# if we want to actually just search for a period, we have to escape it.
pattern = re.compile(r'\.')
matches = pattern.finditer(text_to_search)
for match in matches:
    print(match)
```

```python
# a perhaps practical thing would be to match a url as we would have to escape the period
pattern = re.compile(r'coreyms\.com')
matche = pattern.findall(text_to_search)
print(match)
```

```python
# now let's actually do some actual pattern matching...
# the \d matches any digit between 0-9
pattern = re.compile(r'\d')
matches = pattern.finditer(text_to_search)
for match in matches:
    print(match)
```

```python
# \D matches any charater that is not a digit. Kind of negates \d.
pattern = re.compile(r'\D')
matches = pattern.finditer(text_to_search)
for match in matches:
    print(match)
```

```python
# \w searches for a word character. A word character is any character in a-z, A-Z, 0-9 and an underscore.
pattern = re.compile(r'\w')
matches = pattern.finditer(text_to_search)
for match in matches:
    print(match)
```

```python
# \W as you would probably guess searches for anything that is not a word character.
pattern = re.compile(r'\W')
matches = pattern.finditer(text_to_search)
for match in matches:
    print(match)
```

```python
# \s matches any whitespaces: that means a space, a tab or a newline
pattern = re.compile(r'\s')
matches = pattern.finditer(text_to_search)
for match in matches:
    print(match)
```

```python
# \S matches any pattern that is not a whitespace, basically opposite of the above.
pattern = re.compile(r'\S')
matches = pattern.finditer(text_to_search)
for match in matches:
    print(match)
```

#### Now a look at anchors and what they are....

```python
# Anchors do not actually match any characters but rather envisible positions before or after characters.
# We can use anchors in conjuction with other patterns we are searching for.
# First we have \b which is a word boundary. Word boundaries are indicated by a whitespace or a non-alnum character.
# Say we want to match all the 'ha's that have a word boundary directly before.
# The result matchesthe first one because the start of a new line is a word boundary; and the second because the space
# before it is also a word boundary.
# It's not match the last one in 'HaHa' because there is no word boundary before it.
pattern = re.compile(r'\bHa')
matches = pattern.finditer(text_to_search)
for match in matches:
    print(match)
```

```python
# Now,if we replace the search pattern with \B, we should match that last 'Ha'. 
# \B indicates not a word boundary.
pattern = re.compile(r'\BHa')
matches = pattern.finditer(text_to_search)
for match in matches:
    print(match)
```

```python
# ^ will match a position that is the beginning of a string.
# $ will match a position that is the end of a string.
pattern = re.compile(r'^Start')
matches = pattern.finditer(sentence)
for match in matches:
    print(match)
```

```python
pattern = re.compile(r'end$')
matches = pattern.finditer(sentence)
for match in matches:
    print(match)
```

```python
# Say we wanted to match the phone numbers 
pattern = re.compile(r'\d{3}\.\d{3}\.\d{4}')
matches = pattern.finditer(text_to_search)
for match in matches:
    print(match)
```

```python
pattern = re.compile(r'\d{3}-\d{3}-\d{4}')
matches = pattern.finditer(text_to_search)
for match in matches:
    print(match)
```

```python
# We can match both of them at once like this...
pattern = re.compile(r'\d{3}.\d{3}.\d{4}')
matches = pattern.finditer(text_to_search)
for match in matches:
    print(match)
```

### Regular expressions with files

```python
# let's now read in the data.txt file and try to match some patterns, like the urls and phone numbers.
with open('data.txt', 'r') as file:
    contents = file.read()
    
    pattern = re.compile(r'\d{3}-\d{3}-\d{4}')
    
    matches = pattern.finditer(contents)
    
    for match in matches:
        print(match)
```

```python
## Using character sets
# The search pattern below will return all the phone numbers regardless of the separators between the numbers
pattern = re.compile(r'\d{3}.\d{3}.\d{4}')
matches = pattern.finditer(text_to_search)
for match in matches:
    print(match)
```

```python
# but what if we wanted to return only the numbers where the separator is either an _ or a .
# that is where using a character set comes in...
# you just need to put the characters you need to match inside of square brackets as below! Observe that you wouldn't
# need to escape the dot within a charactser set.
pattern = re.compile(r'\d{3}[.-]\d{3}[.-]\d{4}')
matches = pattern.finditer(text_to_search)
for match in matches:
    print(match)
```

```python
# what if we wanted to match just the numbers starting with either 900 or 800?
pattern = re.compile(r'[89]00[.-]\d{3}[.-]\d{4}')
matches = pattern.finditer(text_to_search)
for match in matches:
    print(match)
```

```python
# let's find the numbers starting with 800 through 999 in the text file.
with open('data.txt', 'r') as file:
    contents = file.read()
    
    pattern = re.compile(r'[98]\d{2}-\d{3}-\d{4}')
    
    matches = pattern.finditer(contents)
    
    for match in matches:
        print(match)
```

```python
# the dash when placed between characters, it can specify a range value.
pattern = re.compile(r'[1-5]')                # all matches will be digits between 1 and 5
matches = pattern.finditer(text_to_search)
for match in matches:
    print(match)
```

```python
# we can do this for letters as well.
pattern = re.compile(r'[a-f]')
matches = pattern.finditer(text_to_search)
for match in matches:
    print(match)
```

```python
# We can actually put these ranges back to back and search multiple ranges.
pattern = re.compile(r'[a-fA-F5-8]')
matches = pattern.finditer(text_to_search)
for match in matches:
    print(match)
```

```python
# Within a character set, the ^ negates the search pattern. 
pattern = re.compile(r'[^a-fA-F5-8]')
matches = pattern.finditer(text_to_search)
for match in matches:
    print(match)
```

```python
# if we wanted to match the words cat, mat, pat but not bat; ie all words ending in 'at' but for bat...
pattern = re.compile(r'[cmp]at')
matches = pattern.finditer(text_to_search)
for match in matches:
    print(match)
```

```python
# now that we know the negation power of the ^, let's be a litle smarter....
pattern = re.compile(r'[^b]at')
matches = pattern.finditer(text_to_search)
for match in matches:
    print(match)
```

### Quantifiers in regular expressions...

```python
# * will match 0 or more specified characters
# + will match 1 or more specified characters
# ? will match 0 or one  specified characters
# {4} will match an exact number of specified characters
# {3,4} will match a range of numbers, (minimum, maximum)

pattern = re.compile(r'M(r|s|rs)\.?\s[A-Z]\w*')
matches = pattern.finditer(text_to_search)
for match in matches:
    print(match)
```

```python
emails = '''
CoreyMSchafer@gmail.com
corey.schafer@university.edu
corey-321-schafer@my-work.net
'''
pattern = re.compile(r'[a-zA-Z0-9.-]+@[a-zA-Z-]+\.[a-zA-Z]+')
matches = pattern.finditer(emails)
for match in matches:
    print(match)
```

### Using Groups

```python
urls = '''
https://www.google.com
http://coreyms.com
htpps://youtube.com
https://www.nasa.gov
'''
pattern = re.compile(r'[a-z]+://(www\.)?(\w+)(\.[a-z]+)')
matches = pattern.finditer(urls)

for match in matches:
    print(match.group(0))
```

```python
pattern = re.compile(r'[a-z]+://(www\.)?(\w+)(\.[a-z]+)')
matches = pattern.finditer(urls)

for match in matches:
    print(match.group(1))
```

```python
pattern = re.compile(r'[a-z]+://(www\.)?(\w+)(\.[a-z]+)')
matches = pattern.finditer(urls)

for match in matches:
    print(match.group(2))
```

```python
pattern = re.compile(r'[a-z]+://(www\.)?(\w+)(\.[a-z]+)')
matches = pattern.finditer(urls)

for match in matches:
    print(match.group(3))
```

```python
pattern = re.compile(r'[a-z]+://(www\.)?(\w+)(\.[a-z]+)')
subbed_urls = pattern.sub(r'\2\3', urls)
print(subbed_urls)
```

### Other methods

```python
# the match method; returns the first match. It only matches things at the beginning of the string.
pattern = re.compile(r'Start')
match = pattern.match(sentence)
print(match)
```

```python
# searches entire string for the pattern and just like match, returns the first occurrence.
pattern = re.compile(r'sentence')
match = pattern.search(sentence)
print(match)
```
